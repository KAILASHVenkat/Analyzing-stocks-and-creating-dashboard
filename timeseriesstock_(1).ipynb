```python
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta

# Define the 5 NSE tickers you want
tickers = [
    'RELIANCE.NS', 'TCS.NS', 'INFY.NS', 'HDFCBANK.NS', 'ICICIBANK.NS'
]

# Define the date range for the past 30 days
end_date = datetime.now()
start_date = end_date - timedelta(days=30)

# Function to download data
def download_data(ticker, start, end):
    data = pd.DataFrame()
    current_start = start

    while current_start < end:
        current_end = min(current_start + timedelta(days=7), end)
        temp_data = yf.download(ticker, start=current_start, end=current_end, interval='1m')
        data = pd.concat([data, temp_data])
        current_start = current_end

    return data

# Dictionary to store the downloaded data
all_data = {}

# Download data for the selected tickers
for ticker in tickers:
    try:
        print(f"Downloading data for {ticker}...")
        all_data[ticker] = download_data(ticker, start_date, end_date)
        # Check if the data is downloaded
        if all_data[ticker].empty:
            print(f"No data available for {ticker}")
    except Exception as e:
        print(f"Error downloading data for {ticker}: {e}")
        continue  # Skip to the next ticker if there's an error

# Print descriptive statistics for each ticker's 'Close' price
for ticker in tickers:
    if ticker in all_data and 'Close' in all_data[ticker].columns:
        print(f"\nDescriptive Statistics for {ticker}:")
        print(all_data[ticker]['Close'].describe())  # Descriptive statistics for 'Close' price
    else:
        print(f"No data to show for {ticker}")
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Define the 5 NSE tickers you want
tickers = [
    'RELIANCE.NS', 'TCS.NS', 'INFY.NS', 'HDFCBANK.NS', 'ICICIBANK.NS'
]

# Define the date range for the past 30 days
end_date = datetime.now()
start_date = end_date - timedelta(days=30)

# Function to download data
def download_data(ticker, start, end):
    data = pd.DataFrame()
    current_start = start

    while current_start < end:
        current_end = min(current_start + timedelta(days=7), end)
        temp_data = yf.download(ticker, start=current_start, end=current_end, interval='1m')
        data = pd.concat([data, temp_data])
        current_start = current_end

    return data

# Dictionary to store the downloaded data
all_data = {}

# Download data for the selected tickers
for ticker in tickers:
    try:
        print(f"Downloading data for {ticker}...")
        all_data[ticker] = download_data(ticker, start_date, end_date)
        if all_data[ticker].empty:
            print(f"No data available for {ticker}")
    except Exception as e:
        print(f"Error downloading data for {ticker}: {e}")
        continue

# Create a combined DataFrame with Close prices
combined_data = pd.concat({ticker: data['Close'] for ticker, data in all_data.items()}, axis=1)

# Calculate the Minute Return for each ticker
for ticker in tickers:
    combined_data[('Minute Return', ticker)] = (combined_data[ticker].ffill() / combined_data[ticker].ffill().shift(1)) - 1

# Plotting the Minute Return for each ticker
for ticker in tickers:
    plt.figure(figsize=(10, 6))
    minute_return_series = combined_data[('Minute Return', ticker)].dropna()
    plt.scatter(range(len(minute_return_series)), minute_return_series, label=f'{ticker} Minute Return', s=5)
    plt.title(f'Minute Return for {ticker}')
    plt.xlabel('Minutes (from 0 to total number of minutes)')
    plt.ylabel('Minute Return')
    plt.legend()
    plt.xlim(0, len(minute_return_series))
    plt.show()
```

```python
for ticker in tickers:
    if ticker in all_data and not all_data[ticker].empty:
        first_price = all_data[ticker]['Close'].iloc[0]
        last_price = all_data[ticker]['Close'].iloc[-1]
        cumulative_return = last_price / first_price
        print(f"Cumulative Return for {ticker}: {cumulative_return}")
    else:
        print(f"No data available for {ticker}")
```
