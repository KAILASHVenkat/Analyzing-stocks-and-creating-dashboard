```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
# Define the ticker symbol
ticker_symbol = 'AAPL'

# Define the time frame
end_date = datetime.now()
start_date = end_date - timedelta(days=30)
```

```python
ticker_symbol = 'AAPL'
end_date = datetime.now()
start_date = end_date - timedelta(days=30)

def download_data(ticker, start, end):
    data = pd.DataFrame()
    current_start = start

    while current_start < end:
        current_end = min(current_start + timedelta(days=7), end)
        temp_data = yf.download(ticker, start=current_start, end=current_end, interval='1m')
        data = pd.concat([data, temp_data])
        current_start = current_end

    return data

data = download_data(ticker_symbol, start_date, end_date)
```

```python
data
```

```python
# Step 6: Plot the Data
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Close'], label='Close Price', color='blue')
plt.title(f'{ticker_symbol} Closing Prices for the Past Month')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid()
plt.show()
```
